With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data for df1
data1 = {
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40]
}

# Create DataFrame
df1 = pd.DataFrame(data1)

# Display DataFrame
print("DataFrame 1:")
print(df1)

# Set style for seaborn
sns.set(style="whitegrid")

# Create a histogram for the Age column
plt.figure(figsize=(10, 6))
sns.histplot(df1['Age'], bins=10, kde=True)
plt.title('Age Distribution - Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Create a box plot for the Age column
plt.figure(figsize=(10, 6))
sns.boxplot(x=df1['Age'])
plt.title('Age Distribution - Box Plot')
plt.xlabel('Age')
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset (you should replace this with your actual dataset loading code)
data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Sales': [500, 600, 750, 900, 800],
    'Expenses': [400, 450, 500, 600, 550]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print("DataFrame:")
print(df)

# Select variables for plotting
years = df['Year']
sales = df['Sales']
expenses = df['Expenses']

# Plotting with Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(years, sales, marker='o', linestyle='-', color='b', label='Sales')
plt.plot(years, expenses, marker='s', linestyle='--', color='r', label='Expenses')
plt.title('Sales and Expenses Over Years')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
